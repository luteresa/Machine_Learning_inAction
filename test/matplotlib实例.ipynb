## 画一副高斯分布图

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

高斯分布函数
$$f(x)=\frac{1}{\sqrt{2π}\sigma}*e^{\frac{-(x-u)^2}{2\sigma^2}} $$

```python
def gaussian_density(x,mu,sigma):
    gaussian_probability = (1.0/(np.sqrt(2*np.pi)*sigma))*np.exp(-np.power(x-mu,2)/(2.0*np.power(sigma,2)))
    return gaussian_probability                                                                                                           
```

```python
x = np.linspace(0,100,1000)
mu = 50
sigma=10
plt.plot(x,gaussian_density(x,mu,sigma),'r-',linewidth=1)
plt.show()
```

```python
def plot_fill(x,mu,sigma,x_min,x_max,y_lim,title):
    y = norm(loc=mu,scale=sigma).pdf(x)
    x_fill = np.linspace(x_min,x_max,1000)
    y_fill = norm(loc=mu,scale=sigma).pdf(x_fill)
    
    plt.plot(x,y)
    plt.fill_between(x_fill,y_fill)
    plt.title(title)
    plt.show()
    return None
```
